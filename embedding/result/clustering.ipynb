```python
# load keyword vector
import json

with open('/home/work/nlp/embedding/result/news_title_128.json', 'r') as f:
    news = json.load(f)

print(news['2018-01-01'] )
```

```python
# load keyword 

with open('/home/work/nlp/keyword/news_keyword.json', 'r') as f:
    keyword = json.load(f)

print(keyword['2018-01-01'] )
```

```python
vector_list = []
for item in news.items():
    vector_list.append(item[1])
```

```python
import pandas as pd

dates = pd.date_range('2015-1-1','2022-10-31',freq='D')
dates
```

```python
cluster = pd.DataFrame()
cluster.index = dates
cluster
```

```python
from sklearn.cluster import DBSCAN
import numpy as np

vector = np.array(vector_list) 
model = DBSCAN(eps=0.25,min_samples=3, metric = "cosine")
# 거리 계산 식으로는 Cosine distance를 이용
result = model.fit_predict(vector)
```

```python
np.unique(result)
```

```python
cluster['result'] = result
cluster.head()
```

```python
for cluster_num in set(result):
    # -1,0은 노이즈 판별이 났거나 클러스터링이 안된 경우
    if(cluster_num == -1 or cluster_num == 0): 
        continue
    else:
        print("cluster num : {}".format(cluster_num))
        temp_df = cluster[cluster['result'] == cluster_num] # cluster num 별로 조회
        for title in temp_df.index:
            print(title) # 제목으로 살펴보자
            start = title
            end = start + timedelta(days=1)
            
            while start.strftime("%Y-%m-%d") not in ks11.index :
                start -= timedelta(days=1)
            
            while end.strftime("%Y-%m-%d") not in ks11.index :
                end += timedelta(days=1)
                
            
            print(ks11['Close'][end.strftime("%Y-%m-%d")] - ks11['Close'][start.strftime("%Y-%m-%d")])
            if len(keyword[title.strftime("%Y-%m-%d")]) == 0:
                print()
                continue
            if keyword[title.strftime("%Y-%m-%d")][0] != "":
                print(keyword[title.strftime("%Y-%m-%d")][0][2])
        print()
```
