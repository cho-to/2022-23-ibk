```python
import gensim
from pprint import pprint
from gensim.models.doc2vec import Doc2Vec, TaggedDocument

d1 = gensim.models.Doc2Vec.load('/home/work/nlp/embedding/news_128.doc2vec')
d2 = gensim.models.Doc2Vec.load('/home/work/nlp/embedding/news_256.doc2vec')
```

```python
from konlpy.tag import Mecab

mecab = Mecab()
print(mecab.morphs('안녕하세요. 이건 뭐지?'))
print(d1.infer_vector(mecab.morphs('안녕하세요.')))
```

```python
from numpy import dot
from numpy.linalg import norm

def cos_sim(A, B):
  return dot(A, B)/(norm(A)*norm(B))
```

```python
import pandas as pd

news_data = pd.read_csv('news_title.csv')
news_data.head()
```

```python
# embedding news title
news_data['128'] = news_data.apply(lambda row: model2.encode(row.news), axis = 1)
news_data['256'] = news_data.apply(lambda row: model.encode(row.news), axis = 1)
```

```python
news_data['128d'] = news_data.apply(lambda row: d1.infer_vector(mecab.morphs(row.news)), axis = 1)
news_data['256d'] = news_data.apply(lambda row: d2.infer_vector(mecab.morphs(row.news)), axis = 1)
```

```python
news_data.head()
```

```python
test_news_data = pd.read_csv('news_title_test.csv')
test_news_data.head()
```

```python
result = pd.DataFrame(columns = ['original','128', '256', '128d', '256d'])
result
```

```python
i = 0
for row in test_news_data.itertuples():
    title = row[2]
    
    embedding_128 = model2.encode(row[2])
    news_data['score_128'] = news_data.apply(lambda x: cos_sim(x['128'], embedding_128), axis=1)
    
    embedding_256 = model.encode(row[2])
    news_data['score_256'] = news_data.apply(lambda x: cos_sim(x['256'], embedding_256), axis=1)
    
    embedding_128d = d1.infer_vector(mecab.morphs(row[2]))
    news_data['score_128d'] = news_data.apply(lambda x: cos_sim(x['128d'], embedding_128d), axis=1)
    
    embedding_256d = d2.infer_vector(mecab.morphs(row[2]))
    news_data['score_256d'] = news_data.apply(lambda x: cos_sim(x['256d'], embedding_256d), axis=1)
    
    result.loc[i] = [row[2], news_data.loc[news_data['score_128'].idxmax()]['news'], news_data.loc[news_data['score_256'].idxmax()]['news'], news_data.loc[news_data['score_128d'].idxmax()]['news'], news_data.loc[news_data['score_256d'].idxmax()]['news']]
    i += 1
```
